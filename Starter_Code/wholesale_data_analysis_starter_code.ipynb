## Part 2: Transform the Data
Do we know that this client spent the more money than client 66037? If not, how would we find out? Transform the data using the steps below to prepare it for analysis.

```python
# Create a column that calculates the subtotal for each line using the unit_price and the qty
df['subtotal'] = df['unit_price'] * df['qty']
df
```

```python
# Create a column for shipping price.
# Assume a shipping price of $7 per pound for orders over 50 pounds and $10 per pound for items 50 pounds or under.
df['shipping'] = df['unit_weight'] * df['qty']
df['shipping'] = df['shipping'].apply(lambda x: x * 7 if x > 50 else x * 10)
df

```

```python
# Create a column for the total price using the subtotal and the shipping price along with a sales tax of 9.25%
df['sub_plus_ship'] = df['subtotal'] + df['shipping']
df['total_price'] = df['sub_plus_ship'] + (df['sub_plus_ship'] * 0.0925)
df
```

```python
# Create a column for the cost of each line using unit cost, qty, and
# shipping price (assume the shipping cost is exactly what is charged to the client).
df['cost'] = df['unit_cost'] * df['qty'] + df['shipping']
df
```

```python
# Create a column for the profit of each line using line cost and line price
df['profit'] = df['total_price'] - df['cost']
df
```

## Part 4: Summarize and Analyze
Use the new columns with confirmed values to find the following information.

```python
# How much did each of the top 5 clients by quantity spend? Check your work from Part 1 for client ids.
for client in top_clients:
    print(client, df[df['client_id'] == client]['total_price'].sum())
```

```python
# Create a summary DataFrame showing the totals for the for the top 5 clients with the following information:
# total units purchased, total shipping price, total revenue, and total profit. 
df_summary = pd.DataFrame(columns=['client_id', 'total_units', 'total_shipping', 'total_revenue', 'total_profit'])
df_summary['client_id'] = top_clients
df_summary['total_units'] = df_summary['client_id'].apply(lambda x: df[df['client_id'] == x]['qty'].sum())  
df_summary['total_shipping'] = df_summary['client_id'].apply(lambda x: df[df['client_id'] == x]['shipping'].sum())
df_summary['total_revenue'] = df_summary['client_id'].apply(lambda x: df[df['client_id'] == x]['total_price'].sum())    
df_summary['total_profit'] = df_summary['client_id'].apply(lambda x: df[df['client_id'] == x]['profit'].sum())  
df_summary
```

```python
# Format the data and rename the columns to names suitable for presentation.
if 'total_units' in df_summary.columns:
    df_summary.rename(columns={'total_units': 'Total Units Purchased'}, inplace=True)
if 'total_shipping' in df_summary.columns:
    df_summary.rename(columns={'total_shipping': 'Total Shipping Price'}, inplace=True)
if 'total_revenue' in df_summary.columns:
    df_summary.rename(columns={'total_revenue': 'Total Revenue'}, inplace=True)
if 'total_profit' in df_summary.columns:
    df_summary.rename(columns={'total_profit': 'Total Profit'}, inplace=True)
df_summary.rename(columns={'client_id': 'Client ID', 'total_units': 'Total Units Purchased', 'total_shipping': 'Total Shipping Price', 'total_revenue': 'Total Revenue', 'total_profit': 'Total Profit'}, inplace=True)
# Define the money columns. 
money_columns = ['Total Shipping Price', 'Total Revenue', 'Total Profit']
for column in money_columns:
        if column in df_summary.columns:
            df_summary[column] = pd.to_numeric(df_summary[column], errors='coerce')
# Define a function that converts a dollar amount to millions.
def currency_format_millions(value):
    if pd.notnull(value):
        return f"${value / 1000000:.2f}M"


# Apply the currency_format_millions function to only the money columns. 
for column in money_columns:
    if column in df_summary.columns:
        df_summary[column] = df_summary[column].apply(currency_format_millions)

# Rename the columns to reflect the change in the money format. 
df_summary.rename(columns={'Total Shipping Price': 'Total Shipping Price (Millions)', 'Total Revenue': 'Total Revenue (Millions)', 'Total Profit': 'Total Profit (Millions)'}, inplace=True)
df_summary
```

```python
# Sort the updated data by "Total Profit (millions)" form highest to lowest and assign the sort to a new DatFrame.
df_summary_sorted = df_summary.sort_values(by='Total Profit (Millions)', ascending=False)
df_summary_sorted
```
